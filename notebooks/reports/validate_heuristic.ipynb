```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from aimanager.utils.utils import make_dir

sns.set(rc={'axes.facecolor':'white', 'figure.facecolor':'white'})

folder = '../../data'
merge_file = 'pilot_random1'
plot_folder = 'plots/validate_heuristic'

make_dir(plot_folder)
```

```python
df = pd.read_csv(os.path.join(folder, f'{merge_file}_player_round.csv'))

w = (
    df['experiment_name'] == 'random_1'
)

df['punish_defect_ratio'] = df['punishment'] / (20-df['contribution'])
```

```python
sns.histplot(data=df, x='punish_defect_ratio', hue='experiment_name', stat='percent', common_norm=False)
plt.xlim(0,2)
```

```python
sns.histplot(data=df, x='punish_defect_ratio', hue='experiment_name', stat='percent', common_norm=False, binwidth=0.25)
plt.xlim(0,2)
```
